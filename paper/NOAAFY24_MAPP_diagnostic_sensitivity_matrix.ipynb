```python
import xarray as xr
import xwmb
import matplotlib.pyplot as plt
```

```python
toSv = 1/1035. * 1e-6

labels = ["a)", "b)", "c)", "d)", "e)", "f)", "g)", "h)", "i)"]
kwargs = {"units":"Sv", "alpha":0.7, "lw":1.5}
```

```python
fig, axes = plt.subplots(3,3,figsize=(11,11))
i = 0
for dt in ["monthly", "daily", "hourly"]:
    for gridname in ["zstr", "natv", "rho2"]:
        wmt = xr.open_dataset(f"data/baltic_wmb_heat_{gridname}_{dt}.nc")
        wmt_mean = wmt.mean('time')*toSv
        
        ax = axes.flatten()[i]
        plt.sca(ax)
        xwmb.stairs( wmt_mean, 'mass_tendency',             label=r"$\partial_{t} \mathcal{M}_{\Omega}$",**kwargs)
        xwmb.stairs(-wmt_mean, 'mass_source',               label=r"$-\mathcal{S}_{\Omega}$",            **kwargs)
        xwmb.stairs(-wmt_mean, 'overturning',               label=r"$-\Psi_{\partial \mathcal{R}}$",     **kwargs)
        xwmb.stairs( wmt_mean, 'boundary_fluxes',           label=r"$\mathcal{G}_{\Omega}^{(Surface)}$", **kwargs)
        xwmb.stairs( wmt_mean, 'diffusion',                 label=r"$\mathcal{G}_{\Omega}^{(Mixing)}$",     **kwargs)
        xwmb.stairs( wmt_mean, 'spurious_numerical_mixing', label=r"$\mathcal{G}_{\Omega}^{(S)}$", **kwargs)
        plt.ylabel(r"potential temperature [$\degree$C]");

        if i==6:
            plt.legend(loc="upper left", handlelength=1.2, labelspacing=0.12)
        if gridname != "zstr":
            plt.ylabel("")
        if dt != "hourly":
            plt.xlabel("")
        plt.grid(True, alpha=0.4)
        plt.title(f"{labels[i]} {dt}-mean {gridname} diagnostics")
        plt.ylim(-2, 22);
        plt.xlim(-0.45, 0.45)
        i+=1
plt.tight_layout()
plt.savefig("figures/NOAAFY24_MAPP_offline_budget_convergence_heat.png", bbox_inches="tight", dpi=200)
```
